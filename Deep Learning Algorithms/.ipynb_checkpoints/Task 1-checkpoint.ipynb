```python
#Import libraries
import pandas as pd
import numpy as np
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
#Import data set using Pandas and split 
dataset = pd.read_excel("KDD_DDoS.xlsx")
ipt_data=dataset.iloc[:, :41] #specifying index_location (iloc)
opt_data=dataset.iloc[:, 41:42]
X_train, X_test, Y_train, Y_test = train_test_split(ipt_data,opt_data,test_size=0.4, random_state=1985)
```

# DNN Model

```python
#Define DNN Model
DNN_model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(41,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
#Compile DNN Model
# Compiling the model
DNN_model.compile(optimizer='adam',
                     loss='mse',
                     metrics=['accuracy'])
#Get Model Summary
DNN_model.summary()

# Convert Pandas data frame to a numpy array
X_train_DNN = np.array(X_train)

print(X_train_DNN.shape)
X_train_DNN = X_train_DNN.reshape(X_train_DNN.shape[0], 1, 41)
print(X_train_DNN.shape)

# fitting the DNN model
DNN_model.fit(X_train, Y_train, epochs=50, batch_size=16)

#Evaluate the Model
print("DNN Model Evaluation: ")
DNN_model.evaluate(X_test, Y_test, batch_size=16)
```

# RNN

```python
#Define RNN Model
RNN_model = keras.Sequential([
    keras.layers.Input(shape=(None, 41)),
    keras.layers.SimpleRNN(64, return_sequences=True),
    keras.layers.SimpleRNN(32, return_sequences=True),
    keras.layers.SimpleRNN(16, return_sequences=True),
    keras.layers.SimpleRNN(8),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
#Compile RNN Model
# Compiling the model
RNN_model.compile(optimizer='adam',
                     loss='mse',
                     metrics=['accuracy'])
#Get Model Summary
RNN_model.summary()

#Convert Pandas data frame to a numpy array
X_train_RNN = np.array(X_train)
print(X_train_RNN.shape)

X_train_RNN = X_train_RNN.reshape(X_train_RNN.shape[0], 1, X_train_RNN.shape[1])
print(X_train_RNN.shape)

#Convert Pandas data frame to a numpy array
X_test_RNN = np.array(X_test)
print(X_test_RNN.shape)

X_test_RNN = X_test_RNN.reshape(X_test_RNN.shape[0], 1, X_test_RNN.shape[1])
print(X_test_RNN.shape)

# fitting the RNN model
RNN_model.fit(X_train_RNN, Y_train, epochs=50, batch_size=16)

#Evaluate the Model
print("RNN Model Evaluation: ")
RNN_model.evaluate(X_test_RNN, Y_test, batch_size=16)
```

# RNN_LSTM

```python
#Define RNN_LSTM Model
RNN_LSTM_model = keras.Sequential([
    layers.LSTM(64, activation='relu', return_sequences=True, input_shape=[None, 41]),
    layers.LSTM(32, activation='relu', return_sequences=True),
    layers.LSTM(16, activation='relu', return_sequences=True),
    layers.LSTM(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
#Compile RNN Model
# Compiling the model
RNN_LSTM_model.compile(optimizer='adam',
                     loss='mse',
                     metrics=['accuracy'])
#Get Model Summary
RNN_LSTM_model.summary()

#Convert Pandas data frame to a numpy array
X_train_RNN_LSTM = np.array(X_train)
print(X_train_RNN_LSTM.shape)

X_train_RNN_LSTM = X_train_RNN_LSTM.reshape(X_train_RNN_LSTM.shape[0], 1, X_train_RNN_LSTM.shape[1])
print(X_train_RNN_LSTM.shape)

#Convert Pandas data frame to a numpy array
X_test_RNN_LSTM = np.array(X_test)
print(X_test_RNN_LSTM.shape)

X_test_RNN_LSTM = X_test_RNN_LSTM.reshape(X_test_RNN_LSTM.shape[0], 1, X_test_RNN_LSTM.shape[1])
print(X_test_RNN_LSTM.shape)

# fitting the RNN model
RNN_LSTM_model.fit(X_train_RNN_LSTM, Y_train, epochs=50, batch_size=16)

#Evaluate the Model
print("RNN Model Evaluation: ")
RNN_LSTM_model.evaluate(X_test_RNN_LSTM, Y_test, batch_size=16)
```

# RNN_GRU

```python
from keras.models import Sequential
from keras.layers import GRU, Dense

# Define the model architecture
RNN_GRU_model = Sequential([
    GRU(units=64, return_sequences=True, input_shape=(None, 41)),
    GRU(units=32, return_sequences=True),
    GRU(units=18, return_sequences=True),
    GRU(units=6),
    Dense(units=1)
])

# Compile the model
RNN_GRU_model.compile(optimizer='adam',
                     loss='mse',
                     metrics=['accuracy'])

#Get Model Summary
RNN_GRU_model.summary()
#Convert Pandas data frame to a numpy array
X_train_RNN_GRU= np.array(X_train)
print(X_train_RNN_GRU.shape)

X_train_RNN_GRU = X_train_RNN_GRU.reshape(X_train_RNN_GRU.shape[0], 1, X_train_RNN_GRU.shape[1])
print(X_train_RNN_GRU.shape)

#Convert Pandas data frame to a numpy array
X_test_RNN_GRU = np.array(X_test)
print(X_test_RNN_GRU.shape)

X_test_RNN_GRU = X_test_RNN_GRU.reshape(X_test_RNN_GRU.shape[0], 1, X_test_RNN_GRU.shape[1])
print(X_test_RNN_GRU.shape)

# fitting the RNN model
RNN_GRU_model.fit(X_train_RNN_GRU, Y_train, epochs=50, batch_size=16)

#Evaluate the Model
print("RNN Model Evaluation: ")
RNN_GRU_model.evaluate(X_test_RNN_GRU, Y_test, batch_size=16)
```

# Get Predictions

```python
# reshape the input data for RNN models
X_test_rnn = np.reshape(X_test, (X_test.shape[0], X_test.shape[1]))

# predict the output for each model
y_pred_DNN = DNN_model.predict(X_test)
X_test_rnn = np.expand_dims(X_test_rnn, axis=1) 
y_pred_RNN = RNN_model.predict(X_test_rnn)
y_pred_RNN_LSTM = RNN_LSTM_model.predict(X_test_rnn)
y_pred_RNN_GRU = RNN_GRU_model.predict(X_test_rnn)


y_pred_DNN = (y_pred_DNN > 0.5).astype(int)
y_pred_RNN = (y_pred_RNN > 0.5).astype(int)
y_pred_RNN_LSTM = (y_pred_RNN_LSTM > 0.5).astype(int)
y_pred_RNN_GRU = (y_pred_RNN_GRU > 0.5).astype(int)
```

# Calculate Performance Metrics

```python
deep_learning_models = ["DNN", "RNN", "RNN_LSTM", "RNN_GRU"]

accuracy_score_list = []
TPR_list = []
TNR_list = []
FPR_list = []

#Get accuracy scores
DNN_accuracy_score = accuracy_score(Y_test, y_pred_DNN)
RNN_accuracy_score = accuracy_score(Y_test, y_pred_RNN)
RNN_LSTM_accuracy_score = accuracy_score(Y_test, y_pred_RNN_LSTM)
RNN_GRU_accuracy_score = accuracy_score(Y_test, y_pred_RNN_GRU)

# add accuracy scores to list 
accuracy_score_list.extend([DNN_accuracy_score,RNN_accuracy_score,RNN_LSTM_accuracy_score, RNN_GRU_accuracy_score])

# Print accuracy score for each model
for deep_learning_model, accuracy_scores in zip(deep_learning_models, accuracy_score_list):
    print(f"{deep_learning_model} Accuracy: {accuracy_scores}")

#Get Classification report for each
DNN_classification_report = classification_report(Y_test, y_pred_DNN, zero_division=1)
# Extract the overall accuracy from the report dictionary
overall_accuracy_DNN = float(DNN_classification_report.split()[-2])

RNN_classification_report = classification_report(Y_test, y_pred_RNN, zero_division=1)
# Extract the overall accuracy from the report dictionary
overall_accuracy_RNN = float(RNN_classification_report.split()[-2])

RNN_LSTM_classification_report = classification_report(Y_test, y_pred_RNN_LSTM, zero_division=1)
# Extract the overall accuracy from the report dictionary
overall_accuracy_RNN_LSTM = float(RNN_LSTM_classification_report.split()[-2])

RNN_GRU_classification_report = classification_report(Y_test, y_pred_RNN_GRU, zero_division=1)
# Extract the overall accuracy from the report dictionary
overall_accuracy_RNN_GRU = float(RNN_GRU_classification_report.split()[-2])

#print Classification reports
print(f"{deep_learning_models[0]}: {DNN_classification_report}")
print("DNN Accuracy from classification Report:", overall_accuracy_DNN)

print(f"{deep_learning_models[1]}: {RNN_classification_report}")
print("RNN Accuracy from classification Report:", overall_accuracy_RNN)

print(f"{deep_learning_models[2]}: {RNN_LSTM_classification_report}")
print("RNN_LSTM Accuracy from classification Report:", overall_accuracy_RNN_LSTM)

print(f"{deep_learning_models[3]}: {RNN_GRU_classification_report}")
print("RNN_GRU Accuracy from classification Report:", overall_accuracy_RNN_GRU)
```

# Calculate Confusion Matrix for DNN, RNN, RNN_LSTM. RNN_GRU

```python
#calculate confusion metrics
cm_DNN = confusion_matrix(Y_test, y_pred_DNN)
cm_RNN = confusion_matrix(Y_test, y_pred_RNN)
cm_RNN_LSTM = confusion_matrix(Y_test, y_pred_RNN_LSTM)
cm_RNN_GRU = confusion_matrix(Y_test, y_pred_RNN_GRU)

#Print confusion metrics 
#print Classification reports
print(f"{deep_learning_models[0]} Confusion Matrix: {cm_DNN}")
print(f"{deep_learning_models[1]} Confusion Matrix: {cm_RNN}")
print(f"{deep_learning_models[2]} Confusion Matrix: {cm_RNN_LSTM}")
print(f"{deep_learning_models[3]} Confusion Matrix: {cm_RNN_GRU}")
```

# Calculate Precison, True Positive,  False Positivie Rate

```python
#import libraries 
import matplotlib.pyplot as plt
#colors to plot graph with 
colors = ['Red', 'Purple', 'Cyan', 'Black', 'Yellow']
#Calculate Precison, True Positive, True Negative, False Positivie Rate, 

#Precision Scores
precision_list=[]
dnn_precision = cm_DNN[1, 1] / (cm_DNN[1, 1]+cm_DNN[0, 1])
rnn_precision = cm_RNN[1, 1] / (cm_RNN[1, 1]+cm_RNN[0, 1])
rnn_lstm_precision = cm_RNN_LSTM[1, 1] / (cm_RNN_LSTM[1, 1]+cm_RNN_LSTM[0, 1])
rnn_gru_precision = cm_RNN_GRU[1, 1] / (cm_RNN_GRU[1, 1]+cm_RNN_GRU[0, 1])


#add all the precision scores to a list
precision_list.extend([dnn_precision, rnn_precision, rnn_lstm_precision, rnn_gru_precision])

# Print accuracy score for each model
print("Accuracy :")
for deep_learning_model, accuracy_scores in zip(deep_learning_models, accuracy_score_list):
    print(f"{deep_learning_model} Accuracy: {accuracy_scores}")
    
print("\n")

#print precision values 
print("Precision :")
for deep_learning_model, accuracy in zip(deep_learning_models, accuracy_score_list):
    print(f'{deep_learning_model}: {accuracy}')
    
print("\n")

#True Positive Rate - TPR
tpr_list=[]
dnn_tpr = cm_DNN[1, 1] / (cm_DNN[1, 1] + cm_DNN[1, 0])
rnn_tpr = cm_RNN[1, 1] / (cm_RNN[1, 1] + cm_RNN[1, 0])
rnn_lstm_tpr = cm_RNN_LSTM[1, 1] / (cm_RNN_LSTM[1, 1] + cm_RNN_LSTM[1, 0])
rnn_gru_tpr = cm_RNN_GRU[1, 1] / (cm_RNN_GRU[1, 1] + cm_RNN_GRU[1, 0])


#add all the tpr scores to a list
tpr_list.extend([dnn_tpr, rnn_tpr, rnn_lstm_tpr, rnn_gru_tpr])

#print tpr values 
print("True Positive Rate :")
for deep_learning_model, true_positive_rate in zip(deep_learning_models, tpr_list):
    print(f'{deep_learning_model}: {true_positive_rate}')

print("\n")

#False Positive Rate - FPR
fpr_list=[]
dnn_fpr = cm_DNN[0, 1] / (cm_DNN[0, 0]+cm_DNN[0, 1])
rnn_fpr = cm_RNN[0, 1] / (cm_RNN[0, 0]+cm_RNN[0, 1])
rnn_lstm_fpr = cm_RNN_LSTM[0, 1] / (cm_RNN_LSTM[0, 0]+cm_RNN_LSTM[0, 1])
rnn_gru_fpr = cm_RNN_GRU[0, 1] / (cm_RNN_GRU[0, 0]+cm_RNN_GRU[0, 1])


#add all the fpr scores to a list
fpr_list.extend([dnn_fpr, rnn_fpr, rnn_lstm_fpr, rnn_gru_fpr])

#print fpr values 
print("False Positive Rate :")
for deep_learning_model, false_positive_rate in zip(deep_learning_models, fpr_list):
    print(f'{deep_learning_model}: {false_positive_rate}')
    
```
