# DML USING RAY

# Import Libraries

```python
#Importing Libraries
import time
import ray
import pandas as pd
import numpy as np
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from keras.models import Sequential
from keras.layers import Dense, LSTM, GRU
from keras.optimizers import Adam
from keras.callbacks import LearningRateScheduler
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

# Define a function to create models 

```python
def create_model(model_type, num_input, num_layers, num_nodes, num_output):
    num_output = int(num_output)
    if model_type == 'DNN':
        model = keras.Sequential()
        model.add(layers.Dense(num_nodes[0], activation='relu', input_shape=(num_input,)))
        for i in range(1, num_layers):
            model.add(layers.Dense(num_nodes[i], activation='relu'))
        model.add(layers.Dense(num_output, activation='sigmoid'))
        return model

    elif model_type == 'RNN':
        model = keras.Sequential()
        model.add(layers.SimpleRNN(num_nodes[0], activation='relu', return_sequences=True, input_shape=(num_input, 1)))
        for i in range(1, num_layers - 1):
            model.add(layers.SimpleRNN(num_nodes[i], activation='relu', return_sequences=True))
        model.add(layers.SimpleRNN(num_nodes[-1], activation='relu'))
        model.add(layers.Dense(num_output, activation='sigmoid'))
        return model

    elif model_type == 'RNN_LSTM':
        model = keras.Sequential()
        model.add(layers.LSTM(num_nodes[0], activation='relu', return_sequences=True, input_shape=(num_input, 1)))
        for i in range(1, num_layers - 1):
            model.add(layers.LSTM(num_nodes[i], activation='relu', return_sequences=True))
        model.add(layers.LSTM(num_nodes[-1], activation='relu'))
        model.add(layers.Dense(num_output, activation='sigmoid'))
        return model

    elif model_type == 'RNN_GRU':
        model = Sequential()
        model.add(GRU(units=num_nodes[0], activation='relu', return_sequences=True, input_shape=(num_input, 1)))
        for i in range(1, num_layers - 1):
            model.add(GRU(units=num_nodes[i], activation='relu', return_sequences=True))
        model.add(GRU(units=num_nodes[-1], activation='relu'))
        model.add(Dense(units=num_output, activation='sigmoid'))
        return model

    else:
        print("Invalid model selected / Invalid parameters, try again")
```

# Create a function to train model in parallel 

```python
@ray.remote(num_cpus=4)
def train_model_in_parallel(x_batch, y_batch, model_type, num_input, num_layers, num_nodes, num_output):
    # Create the model
    model = create_model(model_type, num_input, num_layers, num_nodes, num_output)

    # Compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(x_batch, y_batch, epochs=20, batch_size=15)

    # Get the model weights
    weights = model.get_weights()

    return weights
```

# Create a function to train model sequentially

```python
def train_model_in_sequence(x_batches, y_batches, model_type, num_input, num_layers, num_nodes, num_output):
    # Initialize the model
    model = create_model(model_type, num_input, num_layers, num_nodes, num_output)

    # Compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model sequentially
    for x_batch, y_batch in zip(x_batches, y_batches):
        model.fit(x_batch, y_batch, epochs=20, batch_size=15)

    return model
```

# Create the main function

```python
def main(model_type, num_input, num_layers, num_nodes, num_output):
    # Generate some random numerical data
    dataset = pd.read_excel("../KDD_DDoS.xlsx")
    ipt_data = dataset.iloc[:, :41]  # specifying index_location (iloc)
    opt_data = dataset.iloc[:, 41:42]
    x_train, x_test, y_train, y_test = train_test_split(ipt_data, opt_data, test_size=0.3, random_state=1985)

    # Convert y_train to NumPy array
    y_train = y_train.values

    # Reshape for RNN models
    if model_type in ['RNN', 'RNN_LSTM', 'RNN_GRU']:
        # Reshape for multiple timesteps
        x_train = x_train.to_numpy().reshape(x_train.shape[0], x_train.shape[1], 1)
        y_train = y_train.reshape(y_train.shape[0], y_train.shape[1], 1)

    # Split the data into batches
    x_batches = np.array_split(x_train, 6)
    y_batches = np.array_split(y_train, 6)
    
    # Create Parallel Model
    parallel_model = create_model(model_type, num_input, num_layers, num_nodes, num_output)
    # Compile the parallel model
    parallel_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model in parallel
    startTime = time.time()
    results = [
        train_model_in_parallel.remote(
            x_batch,
            y_batch,
            model_type,
            num_input,
            num_layers,
            num_nodes,
            num_output
        )
        for x_batch, y_batch in zip(x_batches, y_batches)
    ]

    # Combine the weights from each batch and update the model
    weights = ray.get(results)
    print("Total time Parallel: ", time.time() - startTime, "seconds")

    parallel_model.set_weights(np.mean(weights, axis=0))

    # Test the parallel model on some new data
    parallel_loss, parallel_accuracy = parallel_model.evaluate(x_test, y_test)
    print("Trained in Parallel")
    print(f'Test loss: {parallel_loss:.3f}, Test accuracy: {parallel_accuracy:.3f}')
    
    # Create sequential model
    sequential_model = create_model(model_type, num_input, num_layers, num_nodes, num_output)
    # Compile the sequential model
    sequential_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model sequentially with the same number of batches
    startTime = time.time()
    for i in range(6):
                sequential_model.fit(x_batches[i], y_batches[i], epochs=20, batch_size=15)
    print("Total time Sequential: ", time.time() - startTime, "seconds")

    # Test the sequential model on some new data
    sequential_loss, sequential_accuracy = sequential_model.evaluate(x_test, y_test)
    print("Trained in Sequence")
    print(f'Test loss: {sequential_loss:.3f}, Test accuracy: {sequential_accuracy:.3f}')

    # Compare the performance of the models
    print("Trained in Parallel")
    print(f'Test loss: {parallel_loss:.3f}, Test accuracy: {parallel_accuracy:.3f}')

    print("Trained in Sequence")
    print(f'Test loss: {sequential_loss:.3f}, Test accuracy: {sequential_accuracy:.3f}')

    if parallel_accuracy > sequential_accuracy:
        print("The parallel model has a higher test accuracy.")
    else:
        print("The sequential model has a higher test accuracy.")
```
